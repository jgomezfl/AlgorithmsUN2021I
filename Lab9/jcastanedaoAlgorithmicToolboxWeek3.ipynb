##3-1: Money Change

```python
%%writefile ex3_1.py

n = int(input())

c = 0

for i in [10, 5, 1]:
    if n>=i:
        q = n//i
        c = c + q
        n = n%i
        if n == 0:
            print(c)
            quit()
```

##3-2: Maximum Value of the Loot

```python
%%writefile ex3_2.py

n, W = list(map(int, input().split()))
l = []
tv = 0

if W == 0:
    print(0)
    quit()

for _ in range(n):
    v, w = list(map(int, input().split()))
    if v == 0:
        continue
    l.append((v, w))

l.sort(key = lambda x: x[0]/x[1], reverse = True)

for v,w in l:
    if W==0:
        print(tv)
        quit()
    a = min(w, W)
    tv = tv + a*v/w
    W = W - a

print(tv)
```

##3-3: Car Fueling

```python
%%writefile ex3_3.py

d = int(input())
m = int(input())
n = int(input())

stops = list(map(int, input().split()))

stops.append(d)
no_stops = 0
ld = d
cd = 0
for i in range(len(stops) - 1):
    if (ld == 0):
        break
    elif ((stops[i + 1] - stops[i]) > m):
        no_stops = -1
        break
    elif ((cd + m) >= stops[i] and (cd + m) < stops[i + 1]):
        cd = stops[i]
        ld = d - stops[i]
        no_stops +=  1

print(no_stops)
```

##3-5: Collecting Signatures

```python
%%writefile ex3_5.py

n = int(input())

l = []

for e in range(n):
    a, b = list(map(int, input().split()))
    l.append((a,b))

l.sort(key = lambda x: x[1])

i = 0
coord = []

while i < n:
    c = l[i]
    while i < n-1 and c[1]>=l[i+1][0]:
        i += 1

    coord.append(c[1])
    i += 1

print(len(coord))

print(' '.join([str(i) for i in coord]))
```

##3-6: Maximum Number of Prizes

```python
%%writefile ex3_6.py

n = int(input())

W = n
p = []

if n == 1:
    print(1)
    print(1)
    quit()

for i in range(1, n):
    if W>2*i:
        p.append(i)
        W = W - i
    else:
        p.append(W)
        break

print(len(p))
print(' '.join([str(i) for i in p]))
```

##3-7: Maximum Salary

```python
%%writefile ex3_7.py

n = int(input())
l = list(map(int, input().split()))

res = []

while l!=[]:
    max_digit = 0
    for digit in l:
        if int(str(digit)+str(max_digit))>=int(str(max_digit)+str(digit)):
            max_digit = digit
    res.append(max_digit)
    l.remove(max_digit)

print(''.join([str(i) for i in res]))
```
